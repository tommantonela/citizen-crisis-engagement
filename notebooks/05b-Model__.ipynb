### Modelo

* Definir qué modelo vamos a usar.

Paquete de modelos: https://www.statsmodels.org/stable/index.html 

* Documentación de fórmulas: https://www.statsmodels.org/dev/example_formulas.html  

* Fixed effects
    * Dummy variable con el user_id para descartar el efecto de eso.

* Control variables
    * En los papers prueban agregar variables de control de los propios tweets y de los usuarios.

* Interaction effect / cross-effect
    * Variables independientes que se afectan entre si.
    * En este caso podría ser que se consideren las emociones (por ejemplo, resumen de emociones positivas o negativas) o que consideremos los otros engagements.

* Multi-collinearity
    * ``variance_inflation_factor``. Se calcula para las diferentes variables.
    * ``variance_inflation_factor`` expects the presence of a constant in the matrix of explanatory variables. One can use ``add_constant`` from statsmodels to add the required constant to the dataframe before passing its values to the function.
    * Si usamos el ``dmatrices`` la constante se agrega.

* Mean centered
    * Se supone que ayuda. Se le aplica a variables numéricas/continuas.
    * Alternativa al log.

* Comparación de modelos:
    * Log likelihood ratio test
        * Obtener los log likelihood de los modelos y pasarlos por una chi^2.
        * Requiere la diferencia entre los degrees of freedom (df) de los dos modelos que se están comparando. 
        * Solo válido para nested models !! (no sería el caso de todos los modelos, tendríamos varias ramas de acuerdo a si incluimos todas las variables de control o no).
    * Akaike information criterion
        * Parecido al likelihood test, pero no tiene la restricción de nested models.
        * A menor valor, mejor modelo.
        * ``model.aic``.

```python
from scipy.stats.distributions import chi2
# nested models H0 (m0) - H1 (m1) donde m0 is nested con m1 
# model.llf
def compute_log_likelihood_ratio(alt_log_likelihood, alt_df, null_log_likelihood, null_df):
    G = 2 * (null_log_likelihood - alt_log_likelihood)
    p_value = chi2.sf(G, abs(null_df - alt_df)) # esta es la diferencia entre los dfs
    return G, p_value
```

```python
# necesito que también se pongan las que no incluyen retweet a la derecha
def get_no_collinearity(name_part,df_,formulas):

    forms = deque()
    formulas_index = {v.split('~')[1]:k for k,v in formulas.items()}    
    dict_ivf = pd.read_pickle('__ivf__'+ name_part+df_+'.pickle')

    for k,v in dict_ivf.items():
        k_mod = k.split('~')[1]
        if k_mod not in formulas_index:
            continue
        vv = v[v['VIF'] > 5]['variables'].values
        vv = [x for x in vv if not x.startswith('C(') and x != 'Intercept']
        ll = len(vv) 
        print(ll,' __ ',formulas_index[k_mod],'::',vv)
        if ll == 0:
#             print(ll,' __ ',formulas_index[k],'::',k)
#             print(v[v['VIF'] > 5]['variables'].values)
            forms.append(formulas_index[k_mod])
    return forms
```
